```python
#Variaveis de entrada
entradas = [1,7,5]
```

```python
#Pesos iniciais
pesos = [0.8, 0.1, 0]
```

```python
#Função soma
def somaFunction(entradas, pesos):
  s = 0
  for i in range(len(entradas)):
    print(entradas[i],"*", pesos[i], "=", (entradas[i] * pesos[i]))
    s += entradas[i] * pesos[i]
  return s
```

```python
#Executar função soma e trazer o resultado
soma = somaFunction(entradas,pesos)
print('retorno =', soma)
```

```python
#Função de ativação - Step Function
def ativacaoFunction(soma):
  if (soma >= 1):
    return 1
  return 0
```

```python
#Executar função de ativação
resultado = ativacaoFunction(soma)
print('resultado =',resultado)
```
